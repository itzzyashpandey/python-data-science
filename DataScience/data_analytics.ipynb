```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
pd.set_option('display.max_columns',None)#SHOW ALL COLOUMNS
```

```python
df=pd.read_excel('Canada.xlsx',skiprows=20, skipfooter=2,sheet_name='Canada by Citizenship')
```

```python
cols_to_drop=['Type','Coverage','AREA','REG','DEV']#list of unwanted columns
df.drop(columns=cols_to_drop,inplace=True) #droping the columns
years=range(1980,2014)                      #list of years 1980-2013
df.rename({                                 
    'OdName': 'Country', 
    'AreaName': 'Continent',
    'RegName': 'Region',
    'DevName':'Status'
    },axis=1,inplace=True)                      #renaming the coumns to our preference
df.set_index('Country',inplace=True)                #making country as index
df['Total']=df[years].sum(axis=1)                   #adding a new column
df.head(3)                                      #show output first 3 rows
```

```python
df.sort_values(by='Total', inplace = True, ascending=False)
df.head()
```

```python
df.rename({'United Kingdom of Great Britain and Northern Ireland':'UK'},inplace=True)
```

```python
df.head(10)['Total'].plot(kind='pie',figsize=(15,5),title='Top 10 countries')
```

```python
df.head(10)['Total'].plot(kind='bar',figsize=(15,5),title='Top 10 countries')
```
