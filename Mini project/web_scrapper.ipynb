```python
from dputils.scrape import extract_many, get_webpage_data
```

```python
i=7
url = f"https://www.amazon.in/s?k=mobile&i=electronics&page={i}&qid=1665828343&ref=sr_pg_2"
soup = get_webpage_data(url)
url
    
```

```python
target = {
    'tag' : 'div',
    'attrs':{'class':'s-main-slot s-result-list s-search-results sg-row'}
    
}
items = {
    'tag': 'div',
    'attrs':{'class':'s-card-container s-overflow-hidden aok-relative puis-include-content-margin puis s-latency-cf-section s-card-border'}
}
title = {
    'tag': 'h2',
    'attrs':{'class':'a-size-mini a-spacing-none a-color-base s-line-clamp-2'}
}
price={
    'tag':'span',
    'attrs':{'class':'a-price-whole'}
}
rating={
    'tag' : 'span',
    'attrs':{'class':'a-icon-alt'}
}
link = {
    'tag': 'a',
    'attrs':{'class':'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'},
    'output':'href'
}
```

```python
out= extract_many(soup, target=target, items=items, title=title, price=price, rating=rating, link=link)
```

```python
import pandas as pd
df = pd.DataFrame(out)
df
```
